```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
pokemon_df = pd.read_csv('/content/Pokemon.csv')

# Create two DataFrame Grass and Water

grass_df = pokemon_df[pokemon_df['Type 1'] == 'Grass']
water_df = pokemon_df[pokemon_df['Type 1'] == 'Water']

# Create the regression plots for each (Grass and Water)

plt.figure(figsize=(8,6))
sns.regplot(x='Attack', y='Defense', data=grass_df)

plt.figure(figsize=(8,6))
sns.regplot(x='Attack', y='Defense', data=water_df)


# Calculate the Pearson correlation for each DataFrame (variables: Attack and Defense)

correlation_grass = grass_df['Attack'].corr(grass_df['Defense'])
print(f'Pearson Correlation for Grass type Pokemon is : {correlation_grass}')

correlation_water = water_df['Attack'].corr(water_df['Defense'])
print(f'Pearson Correlation for Water type Pokemon is : {correlation_water}')



```
