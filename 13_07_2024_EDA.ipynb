```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('/content/transaction_dataset (1).csv')

# Task 1: Count the occurrences of each unique value in the 'Gender' column (including NaN values)
# Create a bar chart
gender_counts = df['Gender'].value_counts(dropna=False)
plt.bar(gender_counts.index.astype(str), gender_counts, color='skyblue')
plt.title('Count of Transactions for Each Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
# Task 2: Create a horizontal bar chart that shows the top 5 most frequent names
name_counts = df['Name'].value_counts().head(5)
name_counts.plot(kind='barh', color='lightgreen')
plt.title('Top 5 Most Frequent Names')
plt.xlabel('Count')
plt.ylabel('Name')
plt.show()
```

```python
# Task 3 Create a filtered DataFrame that includes Category == 'Clothing' and Gender == 'M'.
# How many rows are there in this filtered DataFrame?

filtered_df = df[(df['Category'] == 'Clothing') & (df['Gender'] == 'M')]
num_rows = len(filtered_df)
print(f"The filtered DataFrame has {num_rows} rows")
```
