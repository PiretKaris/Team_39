```python
# required libraries

import pandas as pd #analysis
import matplotlib.pyplot as plt #vizualization
```

```python
pokemon_df = pd.read_csv('/content/Pokemon (1).csv')
pokemon_df.head(n=10)
```

```python
pokemon_df.sample(5) # random
```

```python
# How many rows and columns are in this data?

rows, columns = pokemon_df.shape
print(f'Number of rows is {rows}, Number of columns is {columns}')
```

```python
# How many unique Pokemon names are in the table?

unique_names = pokemon_df['Name'].nunique()
print(unique_names)
```

```python
# How many Pokemons have Attack greater than 100?

attack_greater_than_100 = pokemon_df[pokemon_df['Attack'] > 100]
print(len(attack_greater_than_100))
```

```python
# How many Pokemons have Defence greater than 100?

defense_greater_than_100 = pokemon_df[pokemon_df['Defense'] > 100]
print(len(defense_greater_than_100))
```

```python
# How many Pokemons have Attack greater than 100 and Defence greater than 100?

attack_defense_greater_than_100 = pokemon_df[(pokemon_df['Attack'] > 100) & (pokemon_df['Defense'] > 100)]
print(len(attack_defense_greater_than_100))
```

```python
# VALUE COUNTS!!! Create a grouped DataFrame that counts the number of Pokemons in each group for Type 1.

type1_pokemons = pokemon_df['Type 1'].value_counts().reset_index()
print(type1_pokemons)
```

```python
# BAR Chart

type1_pokemons.plot(x = 'Type 1', y = 'count', kind = 'bar', color='white', edgecolor='black')
plt.title('Pokemon count by Type 1')
plt.xlabel('Type 1')
plt.ylabel('Count of Pokemons')
```

```python
plt.bar(type1_pokemons['Type 1'], type1_pokemons['count'], color=['thistle','skyblue'], edgecolor='maroon')
plt.xticks(rotation=90)
```

```python
# Which pokemon has the biggest attack?

pokemon_df['Attack'].max() #the highest value of the attack
pokemon_df['Attack'].idxmax() #index of the highest value

highest_attack = pokemon_df['Name'].iloc[pokemon_df['Attack'].idxmax()] #iloc - index location
print('Pokemon with the highest Attack is:', highest_attack )
```

```python
# Which pokemon has the lowest attack?

pokemon_df['Attack'].min() #the lowest value of the attack
pokemon_df['Attack'].idxmin() #index of the lowest value

lowest_attack = pokemon_df['Name'].iloc[pokemon_df['Attack'].idxmin()] #iloc - index location
print('Pokemon with the lowest Attack is:', lowest_attack )
```

```python
#adding a column
pokemon_df['My unique column'] = 'Pokemon'
```

```python
#adding calculated columns
pokemon_df['Attack_Defence'] = pokemon_df['Attack'] + pokemon_df['Defense']
```

```python
# quick math calculations
pokemon_df['Attack'].sum()
pokemon_df['Attack'].min()
pokemon_df['Attack'].max()
pokemon_df['Attack'].mean() # the average
```

```python
pokemon_df['Attack above average'] = pokemon_df['Attack'] > pokemon_df['Attack'].mean()
pokemon_df
```

```python
# removing a column

pokemon_df.shape
pokemon_df.drop(columns=['My unique column'])
pokemon_df.shape
```

# HISTOGRAM

```python
#analyze total for each Pokemon

plt.hist(pokemon_df['Total'], color='green', edgecolor='black', alpha=0.7) #alpha = trabsparency, default = 10 bins
plt.xlabel('Total')
plt.ylabel('Number of Pokemons Frequency')

#avg line
plt.axvline(pokemon_df['Total'].mean(), color='red', linestyle='dashed')

# min line
plt.axvline(pokemon_df['Total'].min(), color='red', linestyle='dashed')

# max line
plt.axvline(pokemon_df['Total'].max(), color='blue', linestyle='dashed')

plt.grid(linestyle='--', alpha=0.7)
```

```python
value = plt.hist(pokemon_df['Total'], bins=3, color='green', edgecolor='black', alpha=0.7)

hist_values = value[0]
bin_edges = value[1]

print('Histogram counts:', hist_values)
print('Bin edges', bin_edges) #the end value is not included
```

```python
# value_counts, how can we calculate the same histogram with value counts?

pokemon_df['Total'].value_counts(bins=[179, 379, 579, 781])

(pokemon_df['Total']>579).sum()
```

# TEAM TASK FOR 03/08

```python
# How many Pokémons are with 'Type 1' == Water as a % of total?

water_pokemon_percentage = (pokemon_df['Type 1'] == 'Water').mean() * 100
print(f"Percentage of Water-type Pokémon: {water_pokemon_percentage:.1f}%")
```

```python
# What is the maximum 'Speed' value? What is the minimum 'Speed' value? What is the difference between max and min 'Speed'?

max_speed = pokemon_df['Speed'].max()
min_speed = pokemon_df['Speed'].min()
speed_difference = max_speed - min_speed
print(f"Maximum Speed: {max_speed}")
print(f"Minimum Speed: {min_speed}")
print(f"Difference between max and min Speed: {speed_difference}")
```

```python
# Filter the DataFrame to include only the Pokémon with 'Speed' >= 80. How many Pokémon meet this criterion?

high_speed_pokemon = pokemon_df[pokemon_df['Speed'] >= 80]
print(f"Number of Pokémon with Speed >= 80: {len(high_speed_pokemon)}")

# Visualize the filtered Pokémon with Speed >= 80
plt.hist(high_speed_pokemon['Speed'], bins=3, color='blue', edgecolor='black', alpha=0.7)
plt.xlabel('Speed')
plt.ylabel('Number of Pokémon')
plt.title('Histogram of Pokémon with Speed >= 80 (3 Bins)')
plt.axvline(high_speed_pokemon['Speed'].mean(), color='red', linestyle='dashed', linewidth=1, label=f'Mean: {high_speed_pokemon["Speed"].mean():.2f}')
plt.legend()
```

```python
# (DIFFICULT) Find Pokémon with the longest name (excluding spaces)? What is this pokemons name?

longest_name_pokemon = pokemon_df.loc[pokemon_df['Name'].str.replace(' ', '').str.len().idxmax()]
print('Pokémon with the longest name (excluding spaces):', longest_name_pokemon)
```
